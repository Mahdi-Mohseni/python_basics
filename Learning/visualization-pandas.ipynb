# Matplotlib

```python
# Import
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(start= -10, stop=20, num= 9) # Make a series between 10 and 20 with equal distance
y = x**2 - 8*x - 35
```

```python
# 1
plt.plot(x, y) # Make plot

# Set configs
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('x^2 - 8x - 35')

plt.show() # This line for showing the plot
```

```python
# Create multiple plots
plt.subplot(1, 2, 1) # it is like (row, columns, order)
plt.plot(x, y, 'red')
plt.title('2 Degree')

plt.subplot(1, 2, 2)
plt.plot(x, x, 'blue')
plt.title('1 Degree')

plt.show()
```

```python
# 2 
# OO way

fig = plt.figure() # Create an object, an empty canvas
axes = fig.add_axes([.1, .1, .8, .8]) # [top, bottom, width, height]
axes.plot(x, y)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([.1, .1, .5, .5])
axes.plot(x, y)
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1, .1, .8, .8])
axes2 = fig.add_axes([.5, .5, .3, .3])
axes1.plot(x, y)
axes2.plot(x, x)
axes1.set_xlabel('x')
plt.show()
```

```python
# 3

fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'red')
axes[1].plot(x, x)

plt.tight_layout() # to show better
plt.show()
```

```python
# Change figure size and dpi
fig, axes = plt.subplots(1, 2, figsize=(10, 5)) 
axes[0].plot(x, y, 'red')
axes[0].set_title("polynomial")

axes[1].plot(x, x)

plt.tight_layout() # to show better
fig.savefig('picture.png', dpi= 300) # to save figure in computer, more dpi = more resolution
plt.show()
```

```python
# Two plot in one figure

fig = plt.figure()

ax = fig.add_axes([.1, .1, .8, .8])

ax.plot(x, x/x + 3, label='y = 3')
ax.plot(x, x, label='y = x')
ax.plot(x, 2*x, label='y = 2x')
ax.plot(x, x+5, label='y = x+5')

ax.legend()
ax.set_title("Compare some functions")

ax.set_xbound(lower=0, upper= 20)
ax.set_ybound(lower=0, upper= 40)

plt.tight_layout()
plt.show()
```
